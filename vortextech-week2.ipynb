```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. LOAD & CLEAN
df = pd.read_csv('players.csv')
df = df.drop_duplicates()
print("Nulls:", df.isnull().sum().sum())

# Sirf numeric columns lo heatmap ke liye
numeric_df = df.select_dtypes(include=[np.number])
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='games_starts', y='minutes_90s', alpha=0.5)
plt.title('Games Started vs Minutes Played')
plt.show()
```

### Insight 1: Game Starts Drive Minutes Played
Players who start more games accumulate significantly more minutes, showing starters get most playing time.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='goals', y='assists', alpha=0.5)
plt.title('Goals vs Assists')
plt.show()
```

### Insight 2: Goals and Assists are Connected
The scatter plot shows a positive relationship. Players who score more goals also tend to have more assists.

```python
avg_goals_pos = df.groupby('position')['goals'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
avg_goals_pos.plot(kind='bar', color='orange')
plt.title('Average Goals by Position')
plt.ylabel('Average Goals')
plt.xticks(rotation=45)
plt.show()
```

### Insight 3: Forwards Score the Most Goals
The bar chart shows that Forward positions have the highest average goals. This matches real football where attackers do most of the scoring.
